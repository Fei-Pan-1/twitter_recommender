```python
df2 = pd.read_csv('/drive/My Drive/cs297/FeiPan_TwitterRecommender_CS297/tweets_by_user.csv')
df2
```

```python
hashtags = {}
# t = re.findall(r"#(\w+)", 'srhf#LOVELIFE bvjwno#GOOUTSIDE vjwieib')
# t = re.findall(r"#(\w+)", '')
# print(t)
for i, row in df2.iterrows():
    hashtag = re.findall(r"#(\w+)", row['text'])
    if len(hashtag) > 0:
      for h in hashtag:
        if h in hashtags:
          hashtags[h] += 1;
        else:
          hashtags[h] = 1;

max_value = max(hashtags, key=hashtags.get)
print(max_value, hashtags[max_value], '\n', hashtags)
```

```python
top_hashtags = sorted(hashtags, key=hashtags.get, reverse=True)[:10]
top_hashtags
# top_hashtags = dict(sorted(hashtags.iteritems(), key=hashtags.__getitem__, reverse=True)[:20])
```
